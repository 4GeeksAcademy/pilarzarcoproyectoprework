#### Exercise 09. Is the average of "Valdemorillo" and "Galapagar" prices the same? (★★☆)

Print the both average prices and then write a conclusion about

Hint: Create a new column called `pps` (price per square) and then analyse the values

```python
# TODO
import pandas as pd
import numpy as np
ds = pd.read_csv('assets/real_estate.csv', sep=';')

    # Creamos una columna nueva llamada "pps" en la que obtenemos el valor del metro cuadrado (price/surface)
ds['pps'] = ds['price'] / ds['surface']

mean_pps_valdemorillo = ds[ds['level5'] == 'Valdemorillo']['pps'].mean()
print("The mean price of Valdemorillo:", mean_pps_valdemorillo)

mean_pps_galapagar = ds[ds['level5'] == 'Galapagar']['pps'].mean()
print("The mean price of Galapagar:", mean_pps_galapagar)

    # Nuestra conclusión
print("The price per square meter in Galpagar is significantly higher than in Valdemorillo, with a difference of €300.")



```

#### Exercise 16. Calculate the sample mean and variance of the variables: price, rooms, surface area and bathrooms (you must use the subset obtained in the question 14) (★★★)

Print both values for each variable

```python
# TODO
import pandas as pd
import numpy as np
ds = pd.read_csv('assets/real_estate.csv', sep=';')

    # Creamos la nueva lista de los 4 municipios con los que queremos trabajar
south_belt_population = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
south_belt = ds[ds['level5'].isin(south_belt_population)]

    # Calculamos el la media de las variables(Price, rooms, surface, bathrooms)
mean_price = south_belt['price'].mean()
mean_rooms = south_belt['rooms'].mean()
mean_surface = south_belt['surface'].mean()
mean_bathrooms = south_belt['bathrooms'].mean()

print("Mean price:", mean_price, f"\nMean rooms:", mean_rooms, f"\nMean surface:", mean_surface, f"\nMean bathrooms:", mean_bathrooms)

    # Calculamos la varianza de las variables(Price, rooms, surface, bathrooms)
mean_price = south_belt['price'].var()
mean_rooms = south_belt['rooms'].var()
mean_surface = south_belt['surface'].var()
mean_bathrooms = south_belt['bathrooms'].var()

print(f"\nMean price:", mean_price, f"\nMean rooms:", mean_rooms, f"\nMean surface:", mean_surface, f"\nMean bathrooms:", mean_bathrooms)

```
